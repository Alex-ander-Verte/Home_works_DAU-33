### Задание 1
Скачайте с [сайта](https://grouplens.org/datasets/movielens/) датасет любого размера. Определите, какому фильму было выставлено больше всего оценок 5.0.


```python
import pandas as pd
```

```python
#читаем файл с рейтингами
df_ratings = pd.read_csv('ml-latest-small/ratings.csv')
df_ratings.head()
```

```python
df_ratings.info()
```

```python
#читаем файл с названими фильмов
df_movies = pd.read_csv('ml-latest-small/movies.csv')
df_movies.head()
```

```python
#отфильтровываем фильмы только с оценкой 5 баллов
movies_5 = df_ratings[ df_ratings['rating'] == 5]
```

```python
#создаём сводную таблицу с количеством 5к для каждого фильма и сортируем по убыванию
pd.pivot_table(movies_5, index='movieId', values='rating', aggfunc='count').sort_values(by='rating', ascending=False)
```

```python
#больше всего 5к получил фильм с ID318 ищем его название в df_movies - это Shawshank Redemption, The (1994)
df_movies[df_movies['movieId'] == 318]
```

### Задание 2
По данным файла power.csv посчитайте суммарное потребление стран Прибалтики (Латвия, Литва и Эстония) категорий 4, 12 и 21 за период с 2005 по 2010 год. Не учитывайте в расчётах отрицательные значения quantity.

```python
df_power = pd.read_csv('Основы_pandas/power.csv')
df_power.head()
```

```python
#Страны Прибалтики
baltic = df_power[(df_power['country'] == 'Latvia') | 
                  (df_power['country'] == 'Lithuania') | 
                  (df_power['country'] == 'Estonia')]
baltic.info()
```

```python
#страны Прибалтики категорий 4, 12 и 21 за период с 2005 по 2010 год с положит. quantity
baltic_filter = baltic[(baltic['quantity'] >= 0) & ((baltic['year'] > 2005) & (baltic['year'] < 2010)) & ((baltic['category'] == 4) | (baltic['category'] == 12) | (baltic['category'] == 21))]
baltic_filter.info()
```

```python
#суммарное потребление стран
baltic_filter['quantity'].sum()
```
