Исходные данные имеют следующую структуру:

```

2.    перечень полок, на которых хранятся документы (если документ есть в documents, то он обязательно должен быть и в directories)
```
directories = {
    '1': ['2207 876234', '11-2'],
    '2': ['10006'],

Общие требования к программе:

*    код должен быть грамотно декомпозирован (каждая функция отвечает за свою конкретную задачу, дублирующийся функционал переиспользуется, а его код не повторяется);

````python
documents = [
    {'type': 'passport', 'number': '2207 876234', 'name': 'Василий Гупкин'},
    {'type': 'invoice', 'number': '11-2', 'name': 'Геннадий Покемонов'},
    {'type': 'insurance', 'number': '10006', 'name': 'Аристарх Павлов'}
]

directories = {
    '1': ['2207 876234', '11-2'],
    '2': ['10006'],
    '3': []
}
````

### Пункт 1. Пользователь по команде "p" может узнать владельца документа по его номеру

Примеры работы:

> Введите команду:

Результат:

````python
def doc_owner ():
    """
    Пользователь может узнать владельца документа по его номеру
    """
    number = input('Введите номер документа: ')
    for i in documents:
        if i['number'] == number:
            result = i['name']
            break
        else:
            result = 'Документ не найден в базе'
    print(result)
````

````python
def doc_loc ():
    """
    Пользователь может по номеру документа узнать на какой полке он хранится
    """
    number = input('Введите номер документа: ')
    for k, v in directories.items():
      if number in v:
        result = 'Документ хранится на полке: ' + k
        break
      else:
        result = 'Документ не найден в базе'
    print(result)
````

### Пункт 3. Пользователь по команде "l" может увидеть полную информацию по всем документам

Пример работы:

> Введите команду:
l

Результат:
```

№: 11-2, тип: invoice, владелец: Геннадий Покемонов, полка хранения: 1
№: 10006, тип: insurance, владелец: Аристарх Павлов, полка хранения: 2
```

````python
def doc_info ():
    """
    Пользователь может увидеть полную информацию по всем документам
    """
    for doc in documents:
        number_doc = doc['number']
        type_doc = doc['type']
        owner_doc = doc['name']
        for i in directories:
            for v in directories[i]:
                if number_doc in v:
                    direct_doc = i
        print(f'№:  {number_doc}, тип: {type_doc}, владелец: {owner_doc}, полка хранения: {direct_doc}')
````

### Пункт 5. Пользователь по команде "ds" может удалить существующую полку из данных (только если она пустая)

````python
def del_loc ():
    number_loc = input('Введите номер полки: ')
    if number_loc not in directories.keys():
        print(f'Такой полки не существует. Текущий перечень полок: {",".join(list(directories.keys()))}')
    elif number_loc in directories.keys() and directories[number_loc] == []:
        del(directories[number_loc])
        print(f'Полка удалена. Текущий перечень полок: {",".join(list(directories.keys()))}')
    else:
        print(f'На полке есть документа, удалите их перед удалением полки. Текущий перечень полок: {",".join(list(directories.keys()))}')
````

````python
#Итоговая версия  обязательной части задания
def main ():
   """
    Пользователь по команде "p" может узнать владельца документа по его номеру
    Пользователь по команде "s" может по номеру документа узнать на какой полке он хранится
    Пользователь по команде "l" может увидеть полную информацию по всем документам
    Пользователь по команде "ads" может добавить новую полку
    Пользователь по команде "ds" может удалить существующую полку из данных (только если она пустая)
    Пользователь по команде "q" останавливает программу
    """
command = input('Введите команду: ')
while command != 'q':
  if command == 'p':
    doc_owner()
    command = input('Введите команду: ')
  elif command == 's':
    doc_loc()
    command = input('Введите команду: ')
  elif command == 'l':
    doc_info()
    command = input('Введите команду: ')
  elif command == 'ads':
    ads_loc()
    command = input('Введите команду: ')
  elif command == 'ds':
    del_loc()
    command = input('Введите команду: ')
  else:
    print('Ошибка. Некорректная команда.')
    command = input('Введите команду: ')
````

````python
def del_doc ():
  """
  Пользователь может удалить документ из данных
  """
  doc_num = input('Введите номер документа: ')
  for i in documents:
      if doc_num == i['number']:
        documents.remove(i)
        for i in directories:
          for v in directories[i]:
            if doc_num in v:
              direct_doc = i
        directories[direct_doc].remove(doc_num)
        text_del = 'Документ удален. \nТекущий список документов: '
      else:
        text_del = 'Документ не найден в базе. \nТекущий список документов:'
  print(text_del)
  doc_info()
````

````python
def doc_reloc ():
  """
  Пользователь может переместить документ с полки на полку
  """
  doc_num = input('Введите номер документа: ')
  doc_new_loc = input('Введите номер полки: ')
  if doc_new_loc not in directories.keys():
    print(f'Такой полки не существует.  \nТекущий список полок: {",".join(list(directories.keys()))}')
  else:
    for i in documents:
      if doc_num == i['number']:
        for i in directories:
          for v in directories[i]:
            if doc_num in v:
              direct_doc = i
        directories[direct_doc].remove(doc_num)
        directories[doc_new_loc].append(doc_num)
        text_reloc = 'Документ перемещен. \nТекущий список документов: '
        break
      else:
        text_reloc = 'Документ не найден в базе. \nТекущий список документов: '
    print(text_reloc)
    doc_info()
````
