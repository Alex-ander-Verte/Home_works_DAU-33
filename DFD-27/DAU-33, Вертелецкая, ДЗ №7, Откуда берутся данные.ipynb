Вывести список основных языков, по каждой стране. Для этого:
* Соедините таблицы country и country_language, при помощи оператора присоединения join;
* Оставьте только те строчки, в которых значение колонки “is_official” имеет значение «True». Выполняем это при помощи оператора where.

```python
df = pd.read_sql(
"""
SELECT
  name, language
FROM
  country c
INNER JOIN
  country_language cl
ON
  c.code = cl.country_code
WHERE
  is_official = True
""",
"postgresql://netology:NetoSQL2019@84.201.153.170:19001/world-db"
)
df.head()
```

"Вывести список основных языков, у каждой страны”, теперь уже средствами pandas

```python
df_country = pd.read_sql(
"""
SELECT
  *
FROM
  country
""",
"postgresql://netology:NetoSQL2019@84.201.153.170:19001/world-db"
)
df_country.head()
```

```python
df_country_language = pd.read_sql(
"""
SELECT
  *
FROM
  country_language
""",
"postgresql://netology:NetoSQL2019@84.201.153.170:19001/world-db"
)
df_country_language.head()
```

```python
df_merged = df_country.merge(df_country_language, left_on='code', right_on='country_code', how='inner')
df_merged.head()
```

```python
df_merged[['name', 'language', 'is_official', ]]
```

```python
filter_df_merged = df_merged['is_official'] == True
df_merged[['name', 'language', 'is_official']].loc[filter_df_merged]
```
