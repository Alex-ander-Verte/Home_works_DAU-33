```python
#Выводим датафрейм с названием города и его населением из таблицы City (9)
df_db = pd.read_sql(
"""
SELECT
  name, population
FROM
  city
WHERE
  (name = 'Ciudad de México' AND country_code = 'MEX') OR name = 'Lublin' OR name = 'Tokyo' OR (name = 'St Petersburg' AND country_code = 'RUS') OR name = 'Brno' OR (name = 'London' AND district = 'England') OR name = 'Cairo' OR name = 'Grodno' OR name = 'Budapest' OR name = 'Sydney' OR name = 'Istanbul';
""",
"postgresql://netology:NetoSQL2019@84.201.153.170:19001/world-db"
)
df_db
```

```python
#Объединяем гугл таблицу и БД (10)
df_merged = df_db.merge(df_sheets, left_on='name', right_on='Название города на английском', how='inner')
df_merged
```

```python
#Группируем по странам, рассчитывая среднюю численность населения в городах, в которых отдыхали
df_merged[['name', 'Страна', 'population']].groupby('Страна').mean()
```
